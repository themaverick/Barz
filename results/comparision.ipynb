```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
zero = pd.read_csv("./data/zeroshot.csv")
finetune = pd.read_csv("./data/finetuned.csv")
```

```python
zero.columns
```

```python
a = [zero.iloc[i, 0] for i in range(10)]
b = [finetune.iloc[i, 0] for i in range(10)]

for i in range(len(a)):
    print(f"{a[i]} | {b[i]}")
```

```python
plt.bar(["zeroshot", "Finetuned"],[zero["factual_correctness(mode=f1)"].mean(), finetune["factual_correctness(mode=f1)"].mean()], color=["#1f77b4", "#ff7f0e"], edgecolor='black')
plt.ylabel("Factual Correctness(f1-score)")
plt.ylim(0, 1) 
plt.title("Factual Correctness (F1) Comparison", fontsize=14)
```

```python
plt.bar(["zeroshot", "Finetuned"],[zero["bleu_score"].mean(), finetune["bleu_score"].mean()], color=["#1f77b4", "#ff7f0e"], edgecolor='black')
plt.ylabel("BLEU-Score")
plt.ylim(0, 1) 
plt.title("BLEU Score Comparision", fontsize=14)
```
